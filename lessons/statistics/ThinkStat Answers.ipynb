```python
#question 2

resp = nsfg.ReadFemResp()

new = resp[resp.totincr==14]
other = resp[resp.totincr<14]

parity = new.parity
other = other.parity
def Cohen(group1, group2):
    diff= group1.mean() - group2.mean()
    
    var1 = group1.var()
    var2 = group2.var()
    pooled = (len(group1)*var1 + len(group2)*var2)/ (len(group1)+len(group2))
    return diff/np.sqrt(pooled)

Cohen(parity, other)
```

```python
#question 3
resp = nsfg.ReadFemResp()
```

```python
def BiasPmf(pmf, label):
    new_pmf = pmf.Copy(label=label)

    for x, p in pmf.Items():
        new_pmf.Mult(x, x)
        
    new_pmf.Normalize()
    return new_pmf
```

```python
pmf = thinkstats2.Pmf(resp.numkdhh, label='actual')
```

```python
thinkplot.Pmf(pmf)
thinkplot.Config(xlabel='Number of children', ylabel='PMF')
```

```python
biased = BiasPmf(pmf, label='bias')
```

```python
thinkplot.PrePlot(2)
thinkplot.Pmfs([pmf,biased])
thinkplot.Config(xlabel='Number of children', ylabel='PMF')
```
